***Seaborn, matplotlib: seaborn uses matplotlib underneath to plot graph***

Distplaot: is used for dirtibution plot, it takes input as an array and plots a curve corresponding to the distribution of points in the array

```python
import seaborn as sns
import matplotlib.pyplot as plt

# sns.displot([0,1,2,3,4,5])
sns.distplot([0,1,2,3,4,5], hist=False)

plt.show()
```

***Visualisation of normal distribution***

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist=False)

plt.show()
```
